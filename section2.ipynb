```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

```python
# d(distance)  =  1 / theta(parralax)
# m - M = -5 + 5 * log(d); m=apparent M = absolute

gaia_data = pd.read_csv('gaia.csv')
theta = gaia_data['parallax']
d = np.zeros(len(theta))
a = np.zeros(len(d))
for i in range(len(d)):
    d[i] = 1 / theta[i]
    a[i] = np.log10(d[i])

apm = gaia_data['mean apparent magnitude']
abm = np.zeros(len(a))
for i in range(len(d)):
    abm[i] = apm[i] + 5*(1-a[i])


```

```python
y = abm
time = gaia_data['time period']
x = np.zeros(len(y))
for i in range(len(x)):
    x[i] = np.log10(time[i])

alpha, beta = np.polyfit(x, y, 1)
print(alpha)
print(beta)
''' 
n = len(time)
sumx = 0; sumy = 0; sumx2 = 0; sumxy = 0; meanx = 0; meany = 0
for i in range(len(x)):
    sumx = sumx + x[i]
    sumy = sumy + y[i]
    sumx2 = sumx2 + x[i]*x[i]
    sumxy = sumxy + x[i]*y[i]

meanx = sumx/n
meany = sumy/n
alpha = (n*sumxy - sumx*sumy)/(n*sumx2 - sumx*sumx)
beta = (sumy - alpha*sumx)/n

print(alpha)
print(beta)
'''
```
